```python
import time
```

```python


def task(name):
    print(f"Start {name}")
    time.sleep(2)          # BLOCKS everything
    print(f"End {name}")

task("A")
task("B")
```

```python
import asyncio
```

```python
async def task(name):
    print(f"Start {name}")
    await asyncio.sleep(2)   # DOES NOT block
    print(f"End {name}")
```

```python
async def main():
    await asyncio.gather(
        task("A"),
        task("B")
    )
```
